```python
#Import Primary Modules:
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library

# use the inline backend to generate the plots within the browser
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt

# check for latest version of Matplotlib
print('Matplotlib version: ', mpl.__version__) # >= 2.0.0
```

```python
df_can = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv")
df_can.head()
```

```python
# print the dimensions of the dataframe
print(df_can.shape)
```

```python
df_can.set_index('Country', inplace=True)

# Let's view the first five elements and see how the dataframe was changed
df_can.head()
```

```python
print('data dimensions:', df_can.shape)
```

```python
years = list(map(str,range(1980,2014)))
```

```python
df_line = df_can[years]
df_line
```

```python
total_immigrants = df_line.sum()
total_immigrants
```

```python
fig , ax = plt.subplots()
ax.plot(total_immigrants)
#Setting up the Title
ax.set_title('Immigrants between 1980 to 2013') 
#Setting up the Labels
ax.set_xlabel('Years')
ax.set_ylabel('Total Immigrants')

#Display the plot
plt.show()
```

```python
# Create a figure with two axes in a row

fig, axs = plt.subplots(1, 2, sharey=True)

#Plotting in first axes - the left one
axs[0].plot(total_immigrants)
axs[0].set_title("Line plot on immigrants")

#Plotting in second axes - the right one
axs[1].scatter(total_immigrants.index, total_immigrants)
axs[1].set_title("Scatter plot on immigrants")

axs[0].set_ylabel("Number of Immigrants")
            
#Adding a Title for the Overall Figure
fig.suptitle('Subplotting Example', fontsize=15)

# Adjust spacing between subplots
fig.tight_layout()


# Show the figure
plt.show()
```

```python
# Create a figure with Four axes - two rows, two columns
fig = plt.figure(figsize=(8,4))

# Add the first subplot (top-left)
axs1 = fig.add_subplot(1, 2, 1)
#Plotting in first axes - the left one
axs1.plot(total_immigrants)
axs1.set_title("Line plot on immigrants")

# Add the second subplot (top-right)
axs2 = fig.add_subplot(1, 2, 2)
#Plotting in second axes - the right one
axs2.barh(total_immigrants.index, total_immigrants) #Notice the use of 'barh' for creating horizontal bar plot
axs2.set_title("Bar plot on immigrants")
            
#Adding a Title for the Overall Figure
fig.suptitle('Subplotting Example', fontsize=15)

# Adjust spacing between subplots
fig.tight_layout()


# Show the figure
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10), sharey=False)
# Line plot
axs[0, 0].plot(total_immigrants, color='blue', linewidth=2)
axs[0, 0].set_title('Line Plot - Total Immigrants', fontsize=12, fontweight='bold')
axs[0, 0].set_xlabel('Year')
axs[0, 0].set_ylabel('Number of Immigrants')
axs[0, 0].grid(True, alpha=0.3)
# Scatter plot
axs[0, 1].scatter(total_immigrants.index, total_immigrants, color='green', alpha=0.6)
axs[0, 1].set_title('Scatter Plot - Total Immigrants', fontsize=12, fontweight='bold')
axs[0, 1].set_xlabel('Year')
axs[0, 1].set_ylabel('Number of Immigrants')
axs[0, 1].grid(True, alpha=0.3)
# Bar plot
axs[1, 0].bar(total_immigrants.index, total_immigrants, color='orange', alpha=0.7)
axs[1, 0].set_title('Bar Plot - Total Immigrants', fontsize=12, fontweight='bold')
axs[1, 0].set_xlabel('Year')
axs[1, 0].set_ylabel('Number of Immigrants')
axs[1, 0].tick_params(axis='x', rotation=45)
axs[1, 0].grid(True, alpha=0.3, axis='y')
# Pie chart
axs[1, 1].pie(df_con['Total'], labels= label , autopct='%1.1f%%', explode=[0,0,0,0,0,0.3])
axs[1, 1].set_title('Pie Chart - Immigrants by Continent', fontsize=12, fontweight='bold')
# Main title    
fig.suptitle('Immigration Data Visualization', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```
