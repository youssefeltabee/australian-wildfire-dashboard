```python
df_can = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv')

print('Data read into a pandas dataframe!')
```

```python
df_can.head(1)
```

```python
df_can.set_index('Country',inplace= True)
df_can.head(1)
```

```python
years = list(map(str,range(1980,2014)))
years
```

```python
# we are using the inline backend
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
print('Matplotlib version: ', mpl.__version__)  # >= 2.0.0
```

```python
print(plt.style.available)
mpl.style.use(['ggplot']) # optional: for ggplot-like style
```

```python
haiti = df_can.loc['Haiti',years]
haiti.head()
```

```python
haiti.plot()
```

```python
haiti.plot(kind = 'line')
plt.title("Immigrants from Haiti")
plt.xlabel("Years")
plt.ylabel("Immigrants Frequency")

plt.show()
```

```python
haiti.index = haiti.index.map(int) 
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

# annotate the 2010 Earthquake. 
# syntax: plt.text(x, y, label)
plt.text(2000, 6000, '2010 Earthquake') # see note below

plt.show() 
```

```python
china_india = df_can.loc[['China','India'],years]
china_india.head()
```

```python
china_india.plot(kind='line')
```

```python
china_india = china_india.transpose()
china_india.head()
```

```python
china_india.plot(kind='line')
plt.title("Immigrants from China and India")
plt.xlabel("Years")
plt.ylabel("Immigrants Frequency")
plt.legend()

plt.show()
```

```python
df_can.sort_values(by='Total',ascending=False,axis=0,inplace=True)
df_can.head()
```

```python
df_can.rename(index={'United Kingdom of Great Britain and Northern Ireland': 'UK'},inplace=True)
```

```python
df_can.head()
```

```python
top_5 = list(df_can.index[0:5])
df_top5 = df_can.loc[top_5,years]
```

```python
df_top5 = df_top5.transpose()
```

```python
df_top5.plot(kind='line')
plt.xlabel("Years")
plt.ylabel('Immigrants Frequency')
plt.legend()
plt.title("Top 5 Countries Immigrated to Canada")
plt.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
# 1. TOP 15 COUNTRIES TREND
top15 = df_can.nlargest(15, 'Total')
fig1 = go.Figure()
for country in top15.index:
    fig1.add_trace(go.Scatter(x=years, y=df_can.loc[country, years], 
                               mode='lines', name=country))
fig1.update_layout(title='Top 15 Countries Immigration Trend',
                   xaxis_title='Year', yaxis_title='Immigrants',
                   height=500)
# 2. IMMIGRATION BY CONTINENT (PIE CHART)
df_cont = df_can.groupby('Continent')['Total'].sum()
fig2 = go.Figure(data=[go.Pie(labels=df_cont.index, values=df_cont.values,
                              hole=0.4, textinfo='label+percent')])
fig2.update_layout(title='Immigration by Continent (Total)')
# 3. HEATMAP - CORRELATION BETWEEN YEARS
fig3 = go.Figure(data=go.Heatmap(z=df_can[years].T.corr(),
                                  x=df_can[years].columns,
                                  y=df_can[years].index,
                                  colorscale='Viridis'))
fig3.update_layout(title='Year-to-Year Correlation Heatmap')
# 4. STACKED BAR - BY REGION (SELECTED YEARS)
df_region = df_can.groupby('Region')[['2000','2005','2010','2013']].sum()
fig4 = go.Figure()
for year in ['2000','2005','2010','2013']:
    fig4.add_trace(go.Bar(x=df_region.index, y=df_region[year], name=year))
fig4.update_layout(barmode='stack', title='Immigration by Region')
fig1.show()
fig2.show()
```
