# Treemaps and Pivot Charts

## Treemap Example

```python
import pandas as pd
import plotly.express as px

# Sales data
data = {
    'Category': ['Electronics', 'Electronics', 'Electronics', 
                 'Furniture', 'Furniture', 'Furniture', 
                 'Clothing', 'Clothing', 'Clothing'],
    'Subcategory': ['Laptops', 'Smartphones', 'Tablets', 
                   'Chairs', 'Tables', 'Sofas', 
                   'Men', 'Women', 'Kids'],
    'Sales': [120000, 80000, 30000, 50000, 40000, 20000, 70000, 90000, 40000]
}
df = pd.DataFrame(data)

fig = px.treemap(df, path=['Category', 'Subcategory'], values='Sales', title='Sales Data Treemap')
fig.show()
```

## Pivot Chart Example

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
data = pd.DataFrame({
    'Date': np.random.choice(['Q1', 'Q2', 'Q3', 'Q4'], 100),
    'Category': np.random.choice(['Peripherals', 'Software', 'Desktops'], 100),
    'Sales': np.random.randint(1000, 5000, 100)
})

pivot_table = data.pivot_table(index='Date', columns='Category', values='Sales', aggfunc=np.sum)
pivot_table.plot(kind='bar', figsize=(10, 6))
plt.title('Sales by Quarter and Category')
plt.xlabel('Quarter')
plt.ylabel('Sales')
plt.show()
```
